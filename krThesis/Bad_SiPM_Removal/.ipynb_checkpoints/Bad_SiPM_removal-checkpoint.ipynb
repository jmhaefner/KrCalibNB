```python
from time import time
t0 = time()
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np
import os
import json as js
```

```python
from scipy.optimize import curve_fit

def gauss2d(xy, amp, x0, y0, s):
    x, y = xy
    inner = (x - x0)**2 / (2*s**2)
    inner += (y - y0)**2 / (2*s**2)
    return amp * np.exp(-inner)
```

```python
nfiles = 10
runno = '7487' # If you want to see nice, accurate Gaussian fits, use 7479.
               # If you want to understand how hard it is to do these fits for low HV, use 7487 :( 
```

```python
from invisible_cities.database import load_db
DataSiPM    = load_db.DataSiPM('new', run_number = int(runno))
DataSiPMX = np.array(DataSiPM.X)
DataSiPMY = np.array(DataSiPM.Y)
```

```python
def printtime(message = ''):
    print(message, 'Current time =', round(time() - t0, 1))
```

```python
filepath = '/Volumes/NEXT_data/IC_Data/krThesis/sliding_window_events/variable_window_json_updated_20201222_r'+runno+'/'
all_json_filenames = [ filepath+filename for filename in os.listdir(filepath) ]

bad_files = []
for json_filename in all_json_filenames:
    if json_filename.find('.json') == -1:
        bad_files.append(json_filename)

for bad_file in bad_files:
    all_json_filenames.remove(bad_file)

if len(all_json_filenames) > nfiles:
    all_json_filenames = all_json_filenames[:nfiles]
```

```python
all_jsons = []
for json_filename in all_json_filenames:
    f = open(json_filename)
    json = js.load(f)
    f.close()
    all_jsons.append(json)
printtime('End all files.')
```

```python
all_jsons[0].keys()
```

```python
ifile = 0
nevts = len(all_jsons[ifile]['events_sum_sipms'])
badsipms = [778, 843, 1346] #[58, 123, 188, 248, 379, 544, 778, 843, 1041, 1086, 1201, 1259, 1346, 1352, 1395, 1465, 1693]
# For source, see "SiPM noise" message from Adam Para

radius = 40
nsipms = len(all_jsons[ifile]['events_sum_sipms'][0])
```

```python
das, dxs, dys = [], [], []

for i in range(nevts):
    E = all_jsons[ifile]['events_window_energy'][i][-1]
    if E > 200 and E < 600:
        testevt = i
        print('iEvt =', testevt, '; E =', E)
        
        # Mask select sipms which fired, and are not considered "bad"
        sipmmask = np.array(all_jsons[ifile]['events_sum_sipms'][testevt]) != 0
        badmask  = [ i not in badsipms for i in range(nsipms)]
        xdata = DataSiPMX[sipmmask & badmask]
        ydata = DataSiPMY[sipmmask & badmask]
        zdata = np.array(all_jsons[ifile]['events_sum_sipms'][testevt])[sipmmask & badmask]

        # Plot the distribution of SiPMs, with brightness indicating strength
        fig = plt.figure(figsize=(6,6))
        ax = fig.add_subplot(111)
        ax.set_title("SiPM brightness vs XY, event "+str(testevt),fontsize=14)
        ax.set_xlabel("X",fontsize=12)
        ax.set_ylabel("Y",fontsize=12)
        ax.grid(True,linestyle='-',color='0.75')
        ax.scatter(xdata,ydata,s=20,c=zdata, marker = 'o', cmap = cm.jet );
        plt.show()
        
        # Determine the center area, and select those x, y, z values
        ibright = np.argmax(all_jsons[ifile]['events_sum_sipms'][testevt])
        xmax, ymax = DataSiPMX[ibright], DataSiPMY[ibright]
        centermask = ((xdata - xmax)**2 + (ydata - ymax)**2) < radius**2
        xdatac, ydatac, zdatac = xdata[centermask], ydata[centermask], zdata[centermask]
        
        # Plot the bright central region
        fig = plt.figure(figsize=(6,6))
        ax = fig.add_subplot(111)
        ax.set_title("SiPM brightness vs XY, central area, event "+str(testevt),fontsize=14)
        ax.set_xlabel("X",fontsize=12)
        ax.set_ylabel("Y",fontsize=12)
        ax.grid(True,linestyle='-',color='0.75')
        ax.scatter(xdatac,ydatac,s=20,c=zdatac, marker = 'o', cmap = cm.jet );
        plt.show()
        
        dataxy = xdata, ydata
        amp0 = max(zdatac)
        x00, y00 = xmax, ymax
        s0 = radius / 2
        pguess = (amp0, x00, y00, s0)
        print('Parameter guess =', pguess)

        popt, pcov = curve_fit(gauss2d, dataxy, zdata, p0=pguess)
        dA, dX, dY =  pguess[0] - popt[0], pguess[1] - popt[1], pguess[2] - popt[2]
        print('dA =', dA)
        print('dX =', dX)
        print('dY =', dY)
        das.append(dA)
        dxs.append(dX)
        dys.append(dY)
```

```python
plt.hist(das, bins = 20)
plt.show()

plt.hist(dxs, bins = 20)
plt.show()

plt.hist(dys, bins = 20)
plt.show()
```
