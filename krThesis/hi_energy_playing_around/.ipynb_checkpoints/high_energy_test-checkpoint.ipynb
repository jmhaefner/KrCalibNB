```python
def NDigit(i, n):
    ret = str(i)
    while len(ret) < n:
        ret = '0' + ret
    return ret
```

```python
all_hdst_filenames = ['hdst_'+NDigit(i, 4)+'_7472_trigger2_v1.1.0_20190801_bg.h5' for i in range(100)]
all_hdsts = None
for hdst_file in all_hdst_filenames:
    print('Appending', hdst_file)
    
    xhdst = pd.HDFStore(hdst_file, 'r')['DST/Events']
    if all_hdsts is None:
        all_hdsts = xhdst
    else:
        all_hdsts = all_hdsts.append(xhdst)
        
print(all_hdsts)
```
