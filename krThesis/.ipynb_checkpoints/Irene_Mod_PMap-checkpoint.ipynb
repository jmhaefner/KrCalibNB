```python
from invisible_cities.io.pmaps_io import load_pmaps
import numpy as np
import matplotlib.pyplot as plt
```

```python
from time import time as time
t0 = time()
```

```python
pmap_file_5us = '/Volumes/NEXT_data/IC_Data/krThesis/modified_peak_finder/irene_out_5us_width.h5'
pmap_file_25us = '/Volumes/NEXT_data/IC_Data/krThesis/modified_peak_finder/irene_out_25us_width.h5'
```

```python
t0 = time()
pmap_5us = load_pmaps(pmap_file_5us)
pmap_25us = load_pmaps(pmap_file_25us)
print('Time to load pmap =', time() - t0)
```

```python
sorted_keys = np.sort(np.array(list(pmap_5us.keys())))
```

```python
pmap_5us[sorted_keys[0]].s2s[0]
```

```python
nplots = 10
npmts = 12
```

```python
for k in sorted_keys[:nplots]:
    for pmt in range(npmts):
        plt.plot( pmap_5us[k].s2s[0].pmts.all_waveforms[pmt] )
    plt.title(k)
    
    all_pmts = [ pmap_5us[k].s2s[0].pmts.all_waveforms[pmt] for pmt in range(12)]
    summed_pmts = np.sum(all_pmts, axis=0)
    plt.plot(summed_pmts, '--')
    
    plt.show()
```

```python
for k in sorted_keys[:nplots]:
    for pmt in range(npmts):
        plt.plot( pmap_25us[k].s2s[0].pmts.all_waveforms[pmt] )
    plt.title(k)
    
    all_pmts = [ pmap_25us[k].s2s[0].pmts.all_waveforms[pmt] for pmt in range(12)]
    summed_pmts = np.sum(all_pmts, axis=0)
    plt.plot(summed_pmts, '--')
    
    plt.show()
```
