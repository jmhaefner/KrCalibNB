## Load in a single file of json, kdst and compare event by event

```python
from time import time
t0 = time()
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json as js
from glob import glob
```

```python
runno = '7479'
```

```python
def ndigit(j, n):
    j = str(j)
    while len(j) < n:
        j = '0' + j
    return j
```

```python
def printtime(message = ''):
    print(message, 'Current time =', round(time() - t0, 1))
```

```python
nfiles = 10
all_kdst_filenames = [ '/Volumes/NEXT_data/IC_Data/kdst/7479/kdst_'+ndigit(i, 4)+'_7479_trigger1_v1.1.0_20190801_autotrigger.h5' for i in range(nfiles) ]
all_json_filenames = [ '/Volumes/NEXT_data/IC_Data/krThesis/sliding_window_events/r7479_d210217_multiple_events_buffer/irene_test_out_'+ndigit(i, 4)+'.json' for i in range(nfiles) ]
```

```python
all_jsons, all_kdsts = [], []
for kdst_filename, json_filename, i in zip(all_kdst_filenames, all_json_filenames, range(len(all_json_filenames))):
    if i % 5 == 0:
        printtime('File '+str(i))
    
    kdst = pd.HDFStore(kdst_filename, 'r')
    f = open(json_filename)
    json = js.load(f)
    f.close()
    
    all_jsons.append(json)
    all_kdsts.append(kdst)
printtime('End all files.')
```

```python
iwindow_ns = 16000
minwindow_ns = 5000
dwindow_ns = 2 * 25 * 5
wideal = int((iwindow_ns - minwindow_ns)/dwindow_ns)
tideal = iwindow_ns
tmax = 25000
```

```python
all_jsons[0].keys()
```

```python
file_evts = []
for json in all_jsons:
    evts = []
    for i in range(len(json['events_window_maxX'])):
        evts.append({'X': json['events_window_maxX'][i][wideal], 'Y' : json['events_window_maxY'][i][wideal], 'Z' : json['events_window_Zgauss'][i][wideal], 'E': (json['events_window_energy'][i][wideal] - (tideal / tmax) * json['events_prewindow_sums'][i])})
    file_evts.append(evts)
```

If we want to get the 16 second cut, with 25 the max, 
dwindow = 5 samples, min = 100, max = 500, then the step
size in terms of ns is 5 x 25 = 125 ns per window step.
That would give a max size of 500 * 125 = 

If the max window is 25000 ns, then 25000 ns / 500 samples
50 ns / sample. So each time, you're increasing by 1 bin RADIUS,
and by 2 bins of diameter. Now, to get 16000 us width, you need
how many samples

The offset is 100 samples = 100 * 50 ns = 5000 ns = 5 us min. Then,
to get the bin, take (16000 ns - 5000 ns) / ((50 ns / step) * (5 step / index))

```python
filepairs = []
dxs, dys, des = [], [], []
xs, ys, zs = [], [], []
es = []
for kdst, json, evts in zip(all_kdsts, all_jsons, file_evts):
    pairs = []
    agree_max = 0
    offset = None
    print('Working on file', len(filepairs), '...')
    for ik in range(len(kdst['DST/Events'].S2e)): # kdst index
        kx, ky, ke = kdst['DST/Events'].X[ik], kdst['DST/Events'].Y[ik], kdst['DST/Events'].S2e[ik]
        
        if ik % 25 == 0:
            print('\t', ik, '/', len(kdst['DST/Events'].S2e))
        # print('\tTrying to match x, y, e =', round(kx,1), round(ky,1), round(ke,1))

        gotpair = False
        
        for ij in range(len(evts)): # json index
            if gotpair or (pairs and ij in np.array(pairs).T[1]):
                continue

            jx, jy, je, jz = evts[ij]['X'], evts[ij]['Y'], evts[ij]['E'], evts[ij]['Z']
            # print('\t\t', ij, '/', len(evts))
            # print('\t\tComparing with x, y, e =', round(jx,1), round(jy,1), round(je,1))                
            
            dx = kx - jx
            dy = ky - jy
            de = ke - je
            
            
            if (abs(dx) < 20) and (abs(dy) < 20) and (abs(de / ke) < 0.15):
                pairs.append((ik, ij))
                # print('\tGOT A MATCH! ik =', ik, '; ij =', ij)
                dxs.append(dx)
                dys.append(dy)
                des.append(de)
                xs.append(jx)
                ys.append(jy)
                zs.append(jz)
                es.append(je)
                gotpair = True
            #else:
            #    print('\t\tNo match from x, y, e =', round(jx), round(jy), round(je), '; dx, dy, de/e =', round(dx), round(dy), round(de / ke))
    printtime('Paired '+str(len(pairs))+'/'+str(len(kdst['DST/Events'].S2e)))
    filepairs.append(pairs)
```

```python
# Want to plot event side by side with json

deas = [] # dE, absolute
deps = [] # dE, percentage

showmax = 20
shown = 0
for pmap, ip, pairs in zip(all_pmaps, range(len(all_pmaps)), filepairs):
    print('\n')
    print('====================================')
    print('====================================')
    print('=========== ip = ', ip,'/', len(all_pmaps)-1, '=============')
    print('====================================')
    print('====================================')

    nmatches = 0
    for i in pmap['PMAPS/S2'].event.unique():

        if i % 20 == 0:
            print(i, '...')

        show = shown < showmax
        shown += 1

        if show:
            print()
            print('==============================')
            print('========== i = ', i, '===========')
            print('==============================')

        y = pmap['PMAPS/S2'][pmap['PMAPS/S2'].event == i].ene
        x = pmap['PMAPS/S2'][pmap['PMAPS/S2'].event == i].time

        if show:
            plt.plot(x, y)
            plt.show()

        evt_indices = np.where(np.array(all_kdsts[ip]['DST/Events'].event) == i)

        if show:
            print('Event', i, 'corresponds to indices', evt_indices[0])

        timemin, timemax = min(x), max(x)

        foundmatch = False
        for evt_index in evt_indices:
            if evt_index in np.array(pairs).T[0]:
                imatch = np.where(np.array(pairs).T[0] == evt_index)[0][0]
                ij = pairs[imatch][1]
                foundmatch = True
                if show:
                    print('Found a match for this event.')
                    print('Matching pair:')
                    print(pairs[imatch])

        if foundmatch and not all_jsons[ip]['events_cutoff'][ij]:

            jx = all_jsons[ip]['events_times'][ij]
            jy = all_jsons[ip]['events_sumwf'][ij]

            if show:
                plt.plot(jx, jy)

                center = (min(jx) + max(jx)) / 2
                jmax, jmin = center + 8000, center - 8000

                plt.axvline(x=timemin, ls='--', color='red')
                plt.axvline(x=timemax, ls='--', color='red')

                plt.axvline(x=jmin, ls='--', color='purple')
                plt.axvline(x=jmax, ls='--', color='purple')

                plt.show()

                print('Json evt', imatch, ' \tS2 energy =', file_evts[ip][ij]['E'])

        if show:
            print('Pmap evt', i, '  \tS2 energy =', sum(y))

        if foundmatch:
            dea = sum(y) - file_evts[ip][ij]['E']
            dep = 100 * dea / file_evts[ip][ij]['E']
            deas.append(dea)
            deps.append(dep)
            nmatches += 1
            if show:
                print('Absolute de =', dea)
                print('Percent de  =', round(dep,1))
                
    print('Number of matches =', nmatches)
            
print('done')
```

```python
plt.hist(deas, range = (-200, 200), bins = 25)
plt.show()

plt.hist(deps, range = (-3, 2), bins = 25)
plt.xlabel('Event by event energy difference (%)')
plt.title('Run 7479, 10 files')
plt.show()
```
