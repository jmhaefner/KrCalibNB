```python
from time import time as time
t0 = time()
```

```python
file_nothr_hiV = '/Volumes/NEXT_data/IC_Data/krThesis/pmaps_0thr_vs_normal/0thr/run_7479_0000_waveforms_irene.h5'
file_nothr_loV = '/Volumes/NEXT_data/IC_Data/krThesis/pmaps_0thr_vs_normal/0thr/run_7487_0000_waveforms_irene.h5'
file_normal_hiV = '/Volumes/NEXT_data/IC_Data/krThesis/pmaps_0thr_vs_normal/normal/run_7479_0000_waveforms_irene.h5'
file_normal_loV = '/Volumes/NEXT_data/IC_Data/krThesis/pmaps_0thr_vs_normal/normal/run_7487_0000_waveforms_irene.h5'
```

```python
t0 = time()
pmap_nothr_hiV = load_pmaps(file_nothr_hiV)
pmap_nothr_loV = load_pmaps(file_nothr_loV)
pmap_normal_hiV = load_pmaps(file_normal_hiV) 
pmap_normal_loV = load_pmaps(file_normal_loV)
print('Time to load pmaps =', time() - t0)
```

```python
Qs = {}
```

```python
pmaps = [ pmap_nothr_hiV, pmap_nothr_loV, pmap_normal_hiV, pmap_normal_loV ]
pmap_names = ['nothr_hiV', 'nothr_loV', 'normal_hiV', 'normal_loV']
```

```python
qualities = ['keys', 'starttimes', 'endtimes', 'widths', 'heights', 'energies']
```

```python
for i in range(len(pmaps)):

    starttimes = [pmaps[i][k].s2s[0].times[0] for k in pmaps[i].keys() ]
    endtimes   = [pmaps[i][k].s2s[0].times[-1] for k in pmaps[i].keys() ]
    widths     = [pmaps[i][k].s2s[0].width for k in pmaps[i].keys() ]
    heights    = [pmaps[i][k].s2s[0].height for k in pmaps[i].keys() ]
    energies   = [pmaps[i][k].s2s[0].total_energy for k in pmaps[i].keys() ]

    Q = {}
    Q['keys'] = pmaps[i].keys()
    Q['starttimes'] = starttimes
    Q['endtimes'] = endtimes
    Q['widths'] = widths
    Q['heights'] = heights
    Q['energies'] = energies
    Qs[pmap_names[i]] = Q
```

```python
Qs.keys()
```

```python
import matplotlib.pyplot as plt
```

```python
for q in qualities:
    print(q)
    for run in pmap_names:
        plt.hist(Qs[run][q])
        plt.title(run+' '+q)
        plt.show()
```

### Check the events in nothr_hiV vs normal_hiV (0, 2)

```python
i_nothr_hiV = 0
i_normal_hiV = 2
thr_killed = []

for i in pmaps[i_nothr_hiV].keys():
    if i not in pmaps[i_normal_hiV].keys():
        thr_killed.append(i)

max_n_plots = 10
n_plots = 0
for i in thr_killed:
    if n_plots >= max_n_plots:
        continue
    n_plots += 1
    for pmt in range(12):
        plt.plot( pmaps[i_nothr_hiV][i].s2s[0].pmts.all_waveforms[pmt] )
    plt.title(i)
    
    all_pmts = [ pmaps[i_nothr_hiV][i].s2s[0].pmts.all_waveforms[pmt] for pmt in range(12)]
    summed_pmts = np.sum(all_pmts, axis=0)
    plt.plot(summed_pmts, '--')
    
    plt.show()
```
